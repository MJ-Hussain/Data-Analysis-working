1. **.isnull()**
2. **.notnull()**

The output is a boolean value.

```python
missing_data=df.isnull()
missing_data.head()
```

`True` means the value is a missing value while `Flase` means the value is not a missing value.

**Mean value for "normalized-losses" column**

As we have seen that the data type for normalized-losses is object. So firstly we have to convert it into float then calculate mean.

```python
avg_norm_loss=df["normalized-losses"].astype('float').mean(axis=0)
print('Average normalized-losses',avg_norm_loss)
```

**Replace `NaN` with mean value in "normalized-losses" column**

```python
df["normalized-losses"].replace(np.nan,avg_norm_loss, inplace=True)
```

**Calculate the mean value for bore and replace `NaN` in "bore" column**

```python
avg_bore=df['bore'].astype('float').mean(axis=0)
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

**Calculate the mean value for "stroke" and replace `NaN` in "stroke" column**

```python
avg_stroke=df['stroke'].astype('float').mean(axis=0)
df['stroke'].replace(np.nan, avg_stroke, inplace=True)
```

**Calculate the mean value for "horsepower" and replace `NaN` in "horsepower" column**

```python
avg_hp=df['horsepower'].astype('float').mean(axis=0)
df['horsepower'].replace(np.nan, avg_hp, inplace=True)
```

**Calculate the mean value for "peak-rpm" and replace `NaN` in "peak-rpm" column**

```python
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

For **num-of-doors** we can use `value_counts()`

```python
df['num-of-doors'].value_counts()
```

We can see that four doors are most common. so we can replace the missing values with this.

```python
df['num-of-doors'].replace(np.nan,"four", inplace=True)
```

Final drop all the rows that do not have price data.

```python
#drop row with NaN in "price" column
df.dropna(subset=['price'], axis=0,inplace=True)

#reset the index after dropping values
df.reset_index(drop=True, inplace=True)
```

### Data Standardization

It is as the process of transforming data into common format, allowing the researcher to make the meaningful comparison.

*Example:* 

Transform mpg to L/100km:

The fuel consumption in our dataset is represented by mpg(miles per gallon) unit in columns 'city-mpg' and 'highway-mpg'. While in some countries, standard used is L/100km.

We will need to applt **data transformation** to transform mpg into L/100km.
The formula for unit conversion is:

L/100km=235/mpg

Check the data before

```python
df.head()
```

```python
#Convert mpg to L/100km
df['city-L/100km']=235/df['city-mpg']

# check the transformed data
df.head()
```

Similarly we transform **"highway-mpg"** ot **"highway-L/100km"**

```python
#Tranforming highway-mpg column to highway-L/100km
df['highway-mpg']=235/df['highway-mpg']
#renaming transformed 'highway-mpg' column to 'highway-L/100km'
df.rename(columns={'highway-mpg':'highway-L/100km'}, inplace=True)
#Take a look at Dataframe
df.head()
```

### Data Normalization

Normalization is the process of transforming values of several variables into a similar range. Typical normalizations include scaling the variables so the variable average is0, scaling the variable so the variance is 1, or scaling the variable so the variable values range from 0 to 1.

*Example:*

Let's say we want to scale the columns "length", "width" and "height".

**Target:** we would like to normalize those variables so their values ranges from 0 to 1.

**Approach:** we will use min-max normalization (replace orignal value by *orignal value/maximun value*)

```python
#Normalizing 'length', 'width' and 'height' columns
df['length']=df['length']/df['length'].max()
df['width']=df['width']/df['width'].max()
df['height']=df['height']/df['height'].max()
```

### Binning
Binning is a process of transforming continuous numerical variable into discrete categorical 'bins' for grouped analysis.

*Example:*

In our dataset, "horsepower" is real valued variable ranging from 48 to 288 and it has 59 unique values. If we only care about the price diggerence between cars with high, medium and low horsepower then we can rearrange then into three **bins** to simplify analysis.

As we want 3 bins, we use `numpy linspace` and divide minimum and maximum values in 4 points divider.

```python
df['horsepower']
```

```python
bins=np.linspace(min(df['horsepower']),max(df['horsepower']), 4)
#bin boundaries
bins
```

we set group names for bins

```python
bin_groups=['low', 'medium', 'high']
```

We apply function `cut`

```python
df['horsepower-binned']=pd.cut(df['horsepower'], bins, labels=bin_groups, include_lowest=True)
#take a look at bins
df[['horsepower','horsepower-binned']].head(10)
```

Let's check the `value_counts()`

```python
df['horsepower-binned'].value_counts()
```

Now plot the distribution of of bins

```python
plt.bar(bin_groups, df['horsepower-binned'].value_counts())
plt.xlabel('horsepower')
plt.ylabel('count')
plt.title('horsepower bins')
```

### Indicator Variable (or Dummy Variable)
An indicator or dummy varible is a mumerical value used to label categories. They are called 'dummies' because the numbers themselves don't have inherent meanings.

*Example:*
In 'fuel-type' coulmn we have two unique values: 'gas' or 'diesel'. To use this attribute in regression analysis, we convert 'fuel-type' to indicator variables as regression doesn't understand words, only numbers.

we use pandas `get_dummies` method to assign numerical values to different categories of fuel type.

```python
dummy_variable_1=pd.get_dummies(df['fuel-type'])
#Check the values
dummy_variable_1.head()
```

Change the column names

```python
dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head()
```

Now merge the dataframe 'df' and 'dummy-variable-1' 

```python
df=pd.concat([df, dummy_variable_1], axis=1)

#drop orignal column 'fuel-type' from 'df'
df.drop('fuel-type', axis=1, inplace=True)

df.head()
```

Similary for 'aspiration' column

```python
#get dummies for aspiration column
dummy_variable2=pd.get_dummies(df['aspiration'])
#rename the dummy_variable
dummy_variable2.rename(columns={'std':'aspiration-type-std','turbo':'aspiration-type-turbo'}, inplace=True)
#merge into the dataframe
df=pd.concat([df, dummy_variable2], axis=1)
#drop the aspiration column from dataframe
df.drop('aspiration', axis=1, inplace=True)

df.head()
```

### Grouping

The `groupby` method groups data by different categories. The data is grouped based on one or several variables, and analysis is performed on the individual groups.

For example, If we want to know, on average, which type of drive wheel is most valuable, we can group 'drive-wheels' and then average them.

```python
df_group=df[['drive-wheels','body-style','price']]

df_group=df_group.groupby(['drive-wheels'],as_index=False).mean()
df_group
```

It seems rear-wheel dirve vehicles are, on average, the most expensive, while 4-wheel and front-wheel are approximately the same in price.

We can also groupby multiple variables. For example, let's group by both 'drive-wheels' and 'body-style'. This groups the dataframe by the unique combination of 'drive-wheels' and 'body-style'. 

```python
df_group=df[['drive-wheels','body-style','price']]
df_group_mod=df_group.groupby(['drive-wheels','body-style'], as_index=False).mean()
df_group_mod
```

#### Pivot table

The grouped data is much easier to visualize when it is made into pivot table. A pivot table is like an Excel spreadsheet, with one variable along the column and another along the row. We can conver the dataframe to a pivot table using the method "pivot" to create a pivot table from the groups.

In this case, we will leave the drive-wheels variable as the rows of the table, and pivot body-style to become the columns of the table:

```python
df_group_pivot=df_group_mod.pivot(index='drive-wheels', columns='body-style')
df_group_pivot
```

Often, we won't have data for some of the pivot cells. We can fill these missing cells with the value 0, but any other value could potentially be used as well.

```python
#fill the missing values with zero
df_group_pivot=df_group_pivot.fillna(0)

df_group_pivot
```
